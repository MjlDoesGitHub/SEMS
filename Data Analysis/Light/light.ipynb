```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("UVdata.csv")
```

```python
df.head()
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(x='Timestamp', y='UV Index', data=df)
plt.title('UV Index Over Time')
plt.xlabel('Timestamp')
plt.ylabel('UV Index')
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Timestamp', y='Visible Light Intensity (lux)', data=df)
plt.title('Visible Light Intensity Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Visible Light Intensity (lux)')
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Timestamp', y='UV Index', data=df, label='UV Index')
sns.lineplot(x='Timestamp', y='Visible Light Intensity (lux)', data=df, label='Visible Light Intensity (lux)')
plt.title('UV Index and Visible Light Intensity Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.legend()
plt.show()
```
